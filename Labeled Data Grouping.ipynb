```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('1st_page_ro42_labeled.csv')
```

```python
df['tag'].unique()
```

```python
promo_features = ['price', 'discount', 'product_name', 'product_desc', 'product_qty']
promo_df = df[df['tag'].isin(promo_features)].copy(deep=True)
```

```python
prices_df = promo_df[promo_df['tag'] == 'price'].copy(deep=True)
prices_df['pair'] = 'N/A'
pair_idx = 10
```

```python
from sklearn.metrics.pairwise import euclidean_distances

dist = euclidean_distances(prices_df[['x_center', 'y_center']].values, prices_df[['x_center', 'y_center']].values)

for i in range(dist.shape[0]):
    dist[i,i] = np.max(dist[i])
```

```python
group = 0
for i in range(dist.shape[0]):
    idx_min = np.argmin(dist[i])
    if prices_df.iloc[i, pair_idx] == 'N/A':
        prices_df.iloc[i, pair_idx] = group
        prices_df.iloc[idx_min, pair_idx] = group 
        group += 1
```

```python
prices_df
```

```python
groups = prices_df['pair'].unique()

glabel_index = 0
value_index = 11
tag_index = 9

for group in groups:
    group_df = prices_df[prices_df['pair'] == group]
    
    assert group_df.shape[0] == 2
    
    first_g_label = group_df.iloc[0, glabel_index]
    second_g_label = group_df.iloc[1, glabel_index]
    
    group_df['value'] = group_df.loc[:, 'Char'].str.replace(',', '.')
    group_df['value'] = group_df['value'].astype(float)

    if group_df.iloc[0, value_index] > group_df.iloc[1, value_index]:
        promo_df.loc[second_g_label, 'tag'] = 'promo price'
    else:
        promo_df.loc[first_g_label, 'tag'] = 'promo price'
```

```python
promo_df['group'] = 'N/A'
promo_df
```

```python
promo_df['tag'].unique()
```

```python
promoprice_df = promo_df[promo_df['tag'] == 'promo price']
discount_df = promo_df[promo_df['tag'] == 'discount']
price_df = promo_df[promo_df['tag'] == 'price']
prod_name_df = promo_df[promo_df['tag'] == 'product_name']
prod_desc_df = promo_df[promo_df['tag'] == 'product_desc']
prod_qty_df = promo_df[promo_df['tag'] == 'product_qty']
```

```python
promoprice_df['group'] = promoprice_df['g_labels']
```

```python
dist = euclidean_distances(price_df[['x_center', 'y_center']].values, promoprice_df[['x_center', 'y_center']].values)
price_df['group'] = promoprice_df.iloc[np.argmin(dist, axis=1), -1].values
```

```python
dist = euclidean_distances(discount_df[['x_center', 'y_center']].values, promoprice_df[['x_center', 'y_center']].values)
discount_df['group'] = promoprice_df.iloc[np.argmin(dist, axis=1), -1].values
```

```python
for index, item in promoprice_df.iterrows():
    prod_candidates = prod_name_df[item['y0'] < prod_name_df['y1']]
    dist = euclidean_distances(prod_candidates[['x_center', 'y_center']].values, [item[['x_center', 'y_center']].values])
    g_label = prod_candidates.iloc[np.argmin(dist), 0]
    prod_name_df.loc[prod_name_df['g_labels'] == g_label, 'group'] = item['g_labels']
```

```python
for index, item in prod_desc_df.iterrows():
    name_candidates = prod_name_df[item['y0'] < prod_name_df['y1']]
    dist = euclidean_distances(name_candidates[['x_center', 'y_center']].values, [item[['x_center', 'y_center']].values])
    groups = name_candidates.iloc[np.argmin(dist), -1]
    prod_desc_df.loc[index, 'group'] = groups
```

```python
item = prod_qty_df.iloc[0, :]
name_candidates = prod_name_df[item['y0'] < rod_name_df['y1']]
name_candidates
```

```python
for index, item in prod_qty_df.iterrows():
    name_candidates = prod_name_df[item['y0'] < prod_name_df['y1']]
    dist = euclidean_distances(name_candidates[['x_center', 'y_center']].values, [item[['x_center', 'y_center']].values])
    groups = name_candidates.iloc[np.argmin(dist), -1]
    prod_qty_df.loc[index, 'group'] = groups
```

```python
promo_v2_df = pd.concat([prod_name_df, prod_desc_df, prod_qty_df, price_df, discount_df, promoprice_df], axis=0).sort_values(by=['group', 'y_center'], ascending=[True, False])
```

```python
tag_order_dict = {
    'product_name': 0,
    'product_desc': 1,
    'product_qty': 2,
    'price': 3,
    'discount': 4,
    'promo price': 5
}

promo_v2_df['tag_order'] = promo_v2_df['tag'].map(tag_order_dict)
promo_v2_df.sort_values(by=['group','tag_order','y_center'], ascending=[True, True, False], inplace=True)
```

```python
from functools import reduce

def concatenate(series):
    return reduce(lambda x, y: x + ' ' + y, series)

promos_v2 = promo_v2_df.groupby('group').agg({'Char': concatenate,
                                  'x0': min,
                                  'x1': max,
                                  'y0': min,
                                  'y1': max})
promos_v2
```
